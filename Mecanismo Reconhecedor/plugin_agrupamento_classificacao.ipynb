```
#Recomenda- se a execução deste código no Google Colaboratory ou Jupyter Notebook

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix

# Tendo separado a base em conjuntos de reino e teste na proporção de 75% e 25%, respectivamente,
# foi feita a predição da classe das instâncias utilizando o algoritmo KNN.
# Depois disso, foi calculada a matriz de confusão para avaliar a acurácia do classificador.

base_dados = pd.read_csv('dados.txt')
base_dados.head()
```

**Código para realizar o agrupamento, utilizando o algoritmo KNN**

Ainda não funciona, deve- se tirar os dados textuais da base

```
todos_tipos = base_dados[base_dados.columns[2]].tolist()

sem_tipos = base_dados.drop(columns=base_dados.columns[2])

treino, teste, tipos, resultado_esperado = train_test_split(sem_tipos, todos_tipos, test_size=0.25, train_size=0.75)

vizinhos = KNeighborsClassifier()

vizinhos = vizinhos.fit(treino, tipos)

previsao = vizinhos.predict(teste)

# O resulatdo é analisado através da matriz de confusão
# É um tabela que mostra as frequências de classificação para cada classe do modelo.
#Indica positivos, falso positivos, falsos e etc.

resultado = confusion_matrix(resultado_esperado, previsao)

print("Matriz de confusão do processo de classificação:")
print(resultado)
```
